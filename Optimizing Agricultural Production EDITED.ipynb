# CLUSTERING ANALYSIS

Clustering analysis is used to classify data points into relative groups called clusters. It is an unspervised learning which means that we will not  provide any labels for training the data, instead we will find out the similar data points by following the existing patterns, within the data. To perform clustering, we are going to use the popular "k means clustering analysis method" which can be imported from the sk-learn library. 

```python
#let's find out the similar crops based on their soil and climatic conditions using the clustering analysis.
# The puropose of this analysis is to find out the similar crops that help the farmers in alternative farming process

#import warnings library so that we can aviod warnings
#import our KMeans lib

from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

```

```python
#let's select the values from the numerical columns and suceesfully remove the "labels" column since unsupervised learning does not approve any labelling in the data. 
x=data.loc[:,['N','P','K','temperature','ph','humidity','rainfall']].values
```

```python
#let's print shape
print(x.shape)
```

```python
#let's convert the data into a dataframe
x_data = pd.DataFrame(x)
x_data.head()
```

```python
#next, we will determine the right number of clusters required for separating these data points using the Elbow method. 
#In order to achieve this, we will be plotting the number of clusters and picking the elbow from the curve as the optimum number of clusters to use

plt.rcParams ['figure.figsize'] =(10,4)

wcss=[]
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state =0)
    km.fit(x)                 
    wcss.append(km.inertia_)
                     
 #let's plot the result
plt.plot(range(1, 11),wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No of Clusters') 
plt.ylabel('wcss')
plt.show()                   
                     
                     
```

There are two elbows in the graph above which can be seen at point 3 and point 4 respectively.But we will have to go with the maximum point which is point 4. This means we are going to have four clusters in our dataset.

```python
#next, let's apply the K Means clustering to our dataset.
#To achieve this, we just have to specify the n clusters parameter inside the k means clustering algorithm.

km = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10, random_state =0)
y_means = km.fit_predict(x)

#lets find out the Results
a = data['label']
y_means = pd.DataFrame(y_means)
z = pd.concat([y_means, a], axis = 1)
z = z.rename(columns = {0: 'cluster'})

#let's check the clusters of each Crops
print("cluster1", z[z['cluster'] == 0]['label'].unique())
print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
print("cluster2", z[z['cluster'] == 1]['label'].unique())
print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
print("clster 3", z[z['cluster'] == 2]['label'].unique())
print("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^")
print("cluster 4", z[z['cluster'] == 3]['label'].unique())
```

From the results above, the crops in the same clusters are similar to each other and this infomation will help the farmers choose the alternate crops very easily and maximize their productivity. 

# BUILDING A PREDICTIVE MODEL

In this stage, we are going to build a predictive model that can suggest the best crop according to the soil and the climatic conditions available for a specific region. The goal of this analysis is to improve agricultural productivity. To build this predictive model, we will train our dataset using machine training model that will predict the most suitable crops using the training dataset and eventually test our result.

```python
#We are going to use the logistic regression model
#first, drop the label's column so it won't interfere with our result and then split our Dataset for predictive modelling

y = data['label']
x = data.drop(['label'], axis=1)

print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
#let's split our dataset into two training ad test set respectively for avlidation of our result.

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size = 0.2, random_state = 44)

print("The shape of x_train:",x_train.shape)
print("The shape of x_test:",x_test.shape)
print("The shape of y_train:", y_train.shape)
print(" The shape of y_test:", y_test.shape)
```

```python
#let's create a Predictive model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
#let's evaluate the Model Performance
from sklearn.metrics import classification_report

#let's print the classification Report also
cr = classification_report(y_test, y_pred)
print(cr)
```
